```python
%matplotlib notebook

import numpy as np
import cv2
import matplotlib.pyplot as pl

# Create a VideoCapture object and read from input file
# to capure from camera argument would be 0
vc = cv2.VideoCapture("2017-04-01_night.mp4")

#interactive pylab is on
pl.ion()
```

```python
fig = pl.figure(figsize=(10,15)).add_subplot(111)
fig.axis('off')
if vc.isOpened(): # try to get the first frame
    is_capturing, frame = vc.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)    # makes the blues image look real colored
    imagecontainer = pl.imshow(frame)    
else:
    is_capturing = False

while is_capturing:
    try:    # keyboardInterrupt to stop the script
        is_capturing, frame = vc.read()
        #frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)    # makes the blues image look real colored
        imagecontainer.set_data(frame)
        pl.draw()

        try:    # Avoids a NotImplementedError caused by `plt.pause`
            pl.pause(0)
        except Exception:
            pass
    except KeyboardInterrupt:
        vc.release() 
```

```python
vc = cv2.VideoCapture("2017-04-01_night.mp4")
ax = pl.figure().add_subplot(111)
ax.axis('off')

is_capturing, frame = vc.read()
frame0 = frame.mean(2)   # makes black and white by averaging 3 colors

imagecontainer = ax.imshow(frame0, cmap='gray')    
```

```python
i = 0

while is_capturing:
    i = i + 10
    vc.set(1, i)
    try:    # keyboardInterrupt to stop the script
        is_capturing, frame1 = vc.read()

        imagecontainer.set_data(np.abs(frame0 - frame1.mean(2)))
        pl.draw()
        try:    # Avoids a NotImplementedError caused by `plt.pause`
            pl.pause(0)
        except Exception:
            pass
    except KeyboardInterrupt:
        vc.release() 
```
