```python
import glob
listf = np.array(glob.glob("faces/*"))

allimgs = np.zeros((len(listf),200*139*4))
#print allimgs.shape
for i,f in enumerate(listf):
    im = imread(f)
    allimgs[i] = im[:200,:139,:].flatten()
nc = 5
clf = sklearn.cluster.KMeans(n_clusters=nc).fit(allimgs)
#the labels tell me which cluster 
#each element of the data array belongs to
#print clf.labels_

for i in range(nc):
    fig = pl.figure(figsize=(8,8))
    k=0
    for j,f in enumerate(listf[clf.labels_==i]):
        k=k+1
        ax = fig.add_subplot(8,5,k)
        ax.axis('off')
        im = imread(f)
        ax.imshow(im)
    figmean = pl.figure(figsize=(5,5)).add_subplot(111)   
    figmean.imshow(np.log10(clf.cluster_centers_[i]).reshape(200,139,4))
    #pl.hist(clf.cluster_centers_[i])
    figmean.axis('off')   
```

```python
allimgs = np.zeros((len(listf),200*139))
#print allimgs.shape
for i,f in enumerate(listf):
    im = imread(f)
    allimgs[i] = im[:200,:139,2].flatten()

nc = 5
clf = sklearn.cluster.KMeans(n_clusters=nc).fit(allimgs)
fig = pl.figure(figsize=(10,10))
#print clf.labels_

for i in range(nc):
    fig = pl.figure(figsize=(8,8))
    k=0
    for j,f in enumerate(listf[clf.labels_==i]):
        k=k+1
        ax = fig.add_subplot(8,5,k)
        ax.axis('off')
        im = imread(f)
        ax.imshow(im)
    figmean = pl.figure(figsize=(5,5)).add_subplot(111)   
    figmean.imshow(np.log10(clf.cluster_centers_[i]).reshape(200,139))
    figmean.axis('off')
```
